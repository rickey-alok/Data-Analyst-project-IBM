## Reading csv flie from online github repository

```python
data = pd.read_csv('https://gist.githubusercontent.com/rnirmal/e01acfdaf54a6f9b24e91ba4cae63518/raw/6b589a5c5a851711e20c5eb28f9d54742d1fe2dc/datasets.csv')
data
```

## adding prefix to all column names

```python
data.add_prefix('x_')
```

## Adding suffix to all column names

```python
data.add_suffix('_Y')
```

## Filter data

```python
data[['datasetName']] # Filtering with 1 Column by providing name of the column
```

```python
data[['datasetName','about']] # Filtering with 2 Columns by providing name of the columns
```

### `[Data + column row]` or  `[Data + row column]`

```python
data[['datasetName','about']][0:6] # Filtering with 2 Columns by providing name of the columns, and 6 rows only
```

```python
data[0:6][['datasetName','about']] # Filtering with 2 Columns by providing name of the columns, and 6 rows only
```

## Filter data based on `LOC` and `iLOC`

## `LOC`

```python
# data.loc[rows,columns]
# rows= need to be `indexed`
# columns = needs to be in `string`

data.loc[0:3,'about':'cloud'] 
```

## `iLOC`

```python
# data.iloc[rows,columns]
# rows= needs to be `indexed`
# columns = needs to be `indexed`

data.iloc[0:3,1:3] 
```

## Filter based on category

```python
data.loc[data['categoryName']=='Natural Language']
```
